```python
from db_connection import get_db_connection
from pymysql import Error

# Insert match details
def insert_match_details(connection, match_data_tuple):
    cursor = connection.cursor()
    insert_query = """
    INSERT INTO match_details (match_id, match_descr, match_format, status, venue, series_id)
    VALUES (%s, %s, %s, %s, %s, %s)
    ON DUPLICATE KEY UPDATE
        match_descr = VALUES(match_descr),
        match_format = VALUES(match_format),
        status = VALUES(status),
        venue = VALUES(venue),
        series_id = VALUES(series_id)
    """
    cursor.execute(insert_query, match_data_tuple)


# Insert team details
def insert_team(connection, team_data_tuple):
    cursor = connection.cursor()
    insert_query = """
    INSERT INTO teams (team_id, team_name, match_id)
    VALUES (%s, %s, %s)
    ON DUPLICATE KEY UPDATE
        team_name = VALUES(team_name)
    """
    cursor.execute(insert_query, team_data_tuple)


# Insert innings scores
def insert_innings_score(connection, innings_tuple):
    cursor = connection.cursor()
    insert_query = """
    INSERT INTO innings_scores (team_id, match_id, innings_number, runs, wickets, overs)
    VALUES (%s, %s, %s, %s, %s, %s)
    """
    cursor.execute(insert_query, innings_tuple)


# Main script
if 'matchDetails' in data and isinstance(data['matchDetails'], list):
    matches = data['matchDetails']
    connection = get_db_connection()

    if connection:
        try:
            for match_details_item in matches:
                if 'matchDetailsMap' in match_details_item and 'match' in match_details_item['matchDetailsMap']:
                    match_data = match_details_item['matchDetailsMap']['match'][0]

                    # --- Match details ---
                    match_id = match_data['matchInfo']['matchId']
                    series_id = match_data['matchInfo']['seriesId']
                    match_descr = match_data['matchInfo']['matchDesc']
                    match_format = match_data['matchInfo']['matchFormat']
                    match_status = match_data['matchInfo']['status']
                    venue = match_data['matchInfo']['venueInfo']['ground']

                    insert_match_details(connection, (match_id, match_descr, match_format, match_status, venue, series_id))

                    # --- Team 1 ---
                    team1_id = match_data['matchInfo']['team1']['teamId']
                    team1_name = match_data['matchInfo']['team1']['teamName']
                    insert_team(connection, (team1_id, team1_name, match_id))

                    team1_inngs1 = match_data['matchScore']['team1Score'].get('inngs1', {})
                    insert_innings_score(connection, (team1_id, match_id, 1,
                                                      team1_inngs1.get('runs', 'N/A'),
                                                      team1_inngs1.get('wickets', 'N/A'),
                                                      team1_inngs1.get('overs', 'N/A')))
                    
                    team1_inngs2 = match_data['matchScore']['team1Score'].get('inngs2')
                    if team1_inngs2:
                        insert_innings_score(connection, (team1_id, match_id, 2,
                                                          team1_inngs2.get('runs', 'N/A'),
                                                          team1_inngs2.get('wickets', 'N/A'),
                                                          team1_inngs2.get('overs', 'N/A')))

                    # --- Team 2 ---
                    team2_id = match_data['matchInfo']['team2']['teamId']
                    team2_name = match_data['matchInfo']['team2']['teamName']
                    insert_team(connection, (team2_id, team2_name, match_id))

                    team2_inngs1 = match_data['matchScore']['team2Score'].get('inngs1', {})
                    insert_innings_score(connection, (team2_id, match_id, 1,
                                                      team2_inngs1.get('runs', 'N/A'),
                                                      team2_inngs1.get('wickets', 'N/A'),
                                                      team2_inngs1.get('overs', 'N/A')))
                    
                    team2_inngs2 = match_data['matchScore']['team2Score'].get('inngs2')
                    if team2_inngs2:
                        insert_innings_score(connection, (team2_id, match_id, 2,
                                                          team2_inngs2.get('runs', 'N/A'),
                                                          team2_inngs2.get('wickets', 'N/A'),
                                                          team2_inngs2.get('overs', 'N/A')))

            # ✅ Commit after all inserts
            connection.commit()
            print("All match, team, and score details inserted successfully.")

        except Error as e:
            print(f"Transaction failed: {e}. Rolling back changes.")
            connection.rollback()
        finally:
            connection.close()
    else:
        print("Failed to connect to database.")
else:
    print("Match data not found or is not a list in the specified structure.")
```

```python
import requests
from db_connection import get_db_connection
from pymysql import Error

# Function to insert venue data into the MySQL table
def insert_venue_data(connection, venue_data):
    cursor = connection.cursor()

    insert_query = """
    INSERT INTO venues (venue_name, city, country, capacity)
    VALUES (%s, %s, %s, %s)
    ON DUPLICATE KEY UPDATE 
        city = VALUES(city),
        country = VALUES(country),
        capacity = VALUES(capacity)
    """
    
    try:
        cursor.execute(insert_query, venue_data)
        connection.commit()
        print(f"Venue inserted/updated successfully: {venue_data[0]}")
    except Error as e:
        print(f"Error while inserting venue data: {e}")
        connection.rollback()


# Function to fetch and insert venue data by venue_id
def process_venue(venue_id, api_key):
    url = f"https://cricbuzz-cricket.p.rapidapi.com/venues/v1/{venue_id}"

    headers = {
        "x-rapidapi-key": api_key,
        "x-rapidapi-host": "cricbuzz-cricket.p.rapidapi.com"
    }

    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        data = response.json()
    else:
        print(f"Failed to fetch venue_id {venue_id}. Status code: {response.status_code}")
        return

    if data:
        connection = get_db_connection()
        if connection:
            venue_name = data.get("ground")
            city = data.get("city")
            country = data.get("country")

            # Clean capacity (remove commas → int)
            raw_capacity = data.get("capacity", "0")
            try:
                capacity = int(raw_capacity.replace(",", "")) if raw_capacity else 0
            except ValueError:
                capacity = 0

            venue_data = (venue_name, city, country, capacity)

            insert_venue_data(connection, venue_data)

            connection.close()
        else:
            print("Failed to connect to the database.")
    else:
        print(f"No data found for venue_id {venue_id}")


# 🚀 Main loop to process multiple venues
if __name__ == "__main__":
    api_key = "4023e3f131msh62f2c03a28eb631p17f1d0jsnce44030e0103"
    
    start_id = 21  # 👈 set your starting venue_id
    end_id = 40    # 👈 set your ending venue_id

    for venue_id in range(start_id, end_id + 1):
        process_venue(venue_id, api_key)
```
